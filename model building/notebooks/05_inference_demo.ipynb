# 05 - Inference Demo

This notebook demonstrates how the backend or dashboard can load the saved models and produce predictions from a new sensor reading.

```python
from pathlib import Path
import sys

cwd = Path.cwd()
if cwd.name == "notebooks":
    BASE_DIR = cwd.parent
elif (cwd / "model building").exists():
    BASE_DIR = cwd / "model building"
else:
    BASE_DIR = cwd

sys.path.append(str(BASE_DIR / "src"))
MODEL_DIR = BASE_DIR / "models"
print("Model directory:", MODEL_DIR)
```

## Sample sensor input

This matches the kind of payload the backend/dashboard can send to the ML layer later.

```python
sample_input = {
    "dust": 35.5,
    "air_quality": 128,
    "temperature": 31.2,
    "humidity": 78.4,
}

sample_input
```

## Run saved models

For demo purposes, rolling and lag features are filled using the current sample. In the real backend, these should be calculated from the latest readings in the same room or session.

```python
from inference import predict_all

result = predict_all(sample_input, model_dir=str(MODEL_DIR))
result
```

## Example output meaning

- `cleanliness_prediction`: learned room cleanliness state.
- `anomaly_status`: whether the reading pattern looks normal or unusual.
- `next_dust_prediction`: short-term forecast for the next dust value.
- `recommendation`: simple dashboard-friendly explanation.
